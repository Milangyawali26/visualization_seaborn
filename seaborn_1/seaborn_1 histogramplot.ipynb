```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore') 

from scipy.stats import norm
```

```python
tips_df=sns.load_dataset("tips")
tips_df.head()
```

```python
sns.distplot(tips_df['size'])
plt.show()
```

```python
sns.distplot(tips_df['total_bill'],)
```

```python
sns.distplot(tips_df['total_bill'],bins=50)
```

```python
sns.distplot(tips_df['total_bill'],bins=50,hist=False)
```

```python
sns.distplot(tips_df['total_bill'],bins=50,kde=False)
```

```python
sns.distplot(tips_df['total_bill'],rug=True)
```

```python
sns.distplot(tips_df['total_bill'],fit=norm,label='total_bill')
plt.legend()
```

```python
# sns.set_style(style="darkgrid")
# #  this is similar to 
sns.set()
sns.distplot(tips_df['total_bill'],fit=norm,label='total_bill')
plt.legend()

plt.title("histogram of total bills")
```

```python
sns.set()
sns.distplot(tips_df['total_bill'],
             hist_kws={'color':'#dc143c'},
             label='total_bill')
plt.legend()

plt.title("histogram of total bills")
```

```python
plt.figure(figsize=(16,9))
sns.distplot(tips_df['total_bill'],label='totalbil')
sns.distplot(tips_df['tip'],label='tip')
sns.distplot(tips_df['size'],bins=9,label='size')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have a DataFrame named tips_df

plt.figure(figsize=(16, 9))

# Apply logarithmic transformation to 'total_bill', 'tip', and 'size'
log_total_bill = np.log(tips_df['total_bill'])
log_tip = np.log(tips_df['tip'])
log_size = np.log(tips_df['size'])

# Plot histograms for log-transformed data
sns.distplot(log_total_bill, label='log_total_bill')
sns.distplot(log_tip, label='log_tip')
sns.distplot(log_size, bins=9, label='log_size')

# Add legend
plt.legend()

# Set plot title and labels
plt.title('Logarithmic Distribution of Total Bill, Tip, and Size')
plt.xlabel('Logarithmic Value')
plt.ylabel('Density')

# Show the plot
plt.show()
```
